```python
# Merge certification_skills into skills
import ast

def merge_cert_skills(row):
    try:
        skills = set(ast.literal_eval(row["skills"])) if pd.notnull(row["skills"]) else set()
        certs = set(ast.literal_eval(row["certification_skills"])) if pd.notnull(row["certification_skills"]) else set()
        return list(skills.union(certs))
    except:
        return row["skills"]

data["skills"] = data.apply(merge_cert_skills, axis=1)
```

```python
def clean_text(txt):
    txt = str(txt).lower()
    txt = re.sub(r'\[.*?\]', '', txt)
    txt = re.sub(r'[^a-z\s]', '', txt)
    return txt.strip()
```

Note: 
str(txt).lower() – Converts to lowercase for case-insensitive comparison (e.g., "Python" == "python").
re.sub(r'\[.*?\]', '', txt) – Removes any unwanted list notation or brackets (like [python, sql]).
re.sub(r'[^a-z\s]', '', txt) – Removes everything except letters and spaces (removes digits, punctuation).
txt.strip() – Trims whitespace from start/end.
Useful to normalize text for matching, vectorization (TF-IDF), or tokenization.

```python
import ast

def safe_join_list(x):
    if isinstance(x, str):
        try:
            items = ast.literal_eval(x)
            return ', '.join(str(i) for i in items if i)  # filters out None and empty strings
        except (ValueError, SyntaxError):
            return ''
    return ''

data['skills'] = data['skills'].apply(safe_join_list)
data['degree_names'] = data['degree_names'].apply(safe_join_list)
```

#### REASON GENERATOR

```python
def generate_reason(row):
    reasons = []
    candidate_skills = set(row['skills'].lower().split(', '))
    required_skills = set(str(row['skills_required']).lower().split(', '))
    if not candidate_skills & required_skills:
        reasons.append("Missing required skills")
    if not any(deg.lower() in str(row['educationaL_requirements']).lower() for deg in row['degree_names'].split(',')):
        reasons.append("Education mismatch")
    if row['experiencere_requirement'].lower() not in str(row['experiencere_requirement']).lower():
        reasons.append("Experience mismatch")
    return ', '.join(reasons) if reasons else 'Profile mismatch'
```

```python
from sklearn.utils.class_weight import compute_class_weight
import numpy as np
from sklearn.metrics import classification_report

# Split RAW TEXT, not TF-IDF
X_text = data['input_text'].values
y = data['shortlisted'].values

X_train_text, X_test_text, y_train, y_test = train_test_split(
    X_text, y, test_size=0.2, random_state=42, stratify=y  # stratify keeps class ratio
)

# Tokenize for RNN
tokenizer = Tokenizer(num_words=10000, oov_token="<OOV>")
tokenizer.fit_on_texts(X_train_text)

X_train_seq = tokenizer.texts_to_sequences(X_train_text)
X_test_seq = tokenizer.texts_to_sequences(X_test_text)

max_len = 300  # tune if needed
X_train_pad = pad_sequences(X_train_seq, maxlen=max_len, padding='post')
X_test_pad = pad_sequences(X_test_seq, maxlen=max_len, padding='post')

# Build LSTM Model
model = Sequential([
    Embedding(input_dim=10000, output_dim=128, input_length=max_len),
    LSTM(128, return_sequences=False),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Compute Class Weights to handle imbalance
class_weights = compute_class_weight(
    class_weight="balanced",
    classes=np.unique(y_train),
    y=y_train
)
class_weights = dict(enumerate(class_weights))
print("Class Weights:", class_weights)

# Train with class weights
model.fit(
    X_train_pad, y_train,
    epochs=10,                
    batch_size=32,
    validation_split=0.1,
    class_weight=class_weights,   
    verbose=1
)

# Evaluate
y_pred_dl = (model.predict(X_test_pad) > 0.5).astype("int32")

print("\nLSTM Model Classification Report:\n")
print(classification_report(y_test, y_pred_dl))
```

#### Tuning 

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout, Bidirectional
from tensorflow.keras.optimizers import Adam

# Build Improved LSTM Model
model = Sequential([
    Embedding(input_dim=10000, output_dim=256, input_length=max_len),  # bigger embeddings
    Bidirectional(LSTM(128, return_sequences=True)),  # bidirectional + return sequences
    Dropout(0.4),
    Bidirectional(LSTM(64, return_sequences=False)),
    Dropout(0.3),
    Dense(128, activation='relu', kernel_regularizer='l2'),
    Dropout(0.3),
    Dense(1, activation='sigmoid')
])

# Compile with lower learning rate
optimizer = Adam(learning_rate=1e-4)
model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])

# Train longer with patience
history = model.fit(
    X_train_pad, y_train,
    epochs=15,               # increased from 5 → 15
    batch_size=32,
    validation_split=0.1,
    callbacks=[
        tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
    ]
)

# Evaluate
y_pred_dl = (model.predict(X_test_pad) > 0.5).astype("int32")
print("\nTuned BiLSTM Model Classification Report:\n")
print(classification_report(y_test, y_pred_dl))
```
